# Extracting the data for more no. of pages
## Here i have taken no. of pages as 3

```python
import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

# Function to scrape data from a given page URL
def scrape_page(link):
    page = requests.get(link)
    soup = bs(page.content, 'html.parser')
    
    products = []
    prices = []
    ratings = []
    storages = []
    sizes = []
    spec_cameras = []
    processors = []

    for data in soup.findAll('div', class_='_3pLy-c row'):
        names = data.find('div', attrs={'class': '_4rR01T'})
        price = data.find('div', attrs={'class': '_30jeq3 _1_WHN1'})
        rating = data.find('div', attrs={'class': '_3LWZlK'})
        specification = data.find('div', attrs={'class': 'fMghEO'})

        for each in specification:
            col = each.find_all('li', attrs={'class': 'rgWa7D'})
            storage = col[0].text.strip()
            size = col[1].text.strip()
            spec_camera = col[2].text.strip()
            processor = col[3].text.strip()
            
        products.append(names.text.strip())
        prices.append(price.text.strip())
        storages.append(storage)
        sizes.append(size)
        spec_cameras.append(spec_camera)
        processors.append(processor)
        ratings.append(rating.text.strip())

    return products, prices, storages, sizes, spec_cameras, processors, ratings

# URL of the first page
base_url = 'https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='

# Scraping data from the first three pages
total_pages = 3
for page_number in range(1, total_pages + 1):
    page_url = base_url + str(page_number)
    products, prices, storages, sizes, spec_cameras, processors, ratings = scrape_page(page_url)

    if page_number == 1:
        # Create the DataFrame for the first page
        df = pd.DataFrame({'Product Name': products, 'Processor': processors, 'Memory': storages,
                           'Length': sizes, 'Specification_of_Camera': spec_cameras, 'Price': prices, 'Rating': ratings})
    else:
        # Append data for subsequent pages to the DataFrame
        df = df.append(pd.DataFrame({'Product Name': products, 'Processor': processors, 'Memory': storages,
                                     'Length': sizes, 'Specification_of_Camera': spec_cameras,
                                     'Price': prices, 'Rating': ratings}))

```

```python
#converting the dataframe to csv

csv_file_path = 'products.csv'
df.to_csv(csv_file_path, index=False)
print(f'DataFrame has been successfully converted to {csv_file_path}')
```
